## Data Preprocessing 

```python
import numpy as np
import pandas as pd
from dotenv import load_dotenv
import os

load_dotenv() # load variables from .env file
```

```python
df = pd.read_csv('bus_data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df[df['bus_type'].isna()]
```

```python
df = df.dropna(subset=['bus_type'])
```

```python
df[df['star_rating'].isna()]
```

```python
df['star_rating'] = pd.to_numeric(df['star_rating'].replace('New', '0'), errors='coerce')
df['star_rating'] = df['star_rating'].fillna(round(df['star_rating'].mean(), 1)).astype(float)
```

```python
df['seats_available'] = df['seats_available'].str.extract('(\d+)').astype(int)
```

```python
df
```

```python
df[df.duplicated()]
```

```python
df = df.drop_duplicates()
```

```python
df.to_csv('final_bus_data.csv', index=False)
```

## Import data into MySQL Database

```python
import pymysql
```

```python
mysql_username = os.getenv("MYSQL_USER")
mysql_password = os.getenv('MYSQL_PASSWORD')
mysql_host = os.getenv("MYSQL_HOST")
mysql_port = os.getenv("MYSQL_PORT")
```

```python
conn = pymysql.connect(
    host=mysql_host,
    user=mysql_username,
    password=mysql_password,
)
cursor = conn.cursor()

# create database if not exists
database_name = 'redbus_database'
cursor.execute(f"CREATE DATABASE IF NOT EXISTS {database_name}")
conn.commit()

# use created or already exists database
cursor.execute(f"USE {database_name}")
```

```python
# create table if it's not exists 
create_table_query = '''
CREATE TABLE IF NOT EXISTS redbus_data (
    id INT AUTO_INCREMENT PRIMARY KEY,
    route_name TEXT,
    route_link TEXT,
    busname TEXT,
    bustype TEXT,
    departing_time TIME,
    duration TEXT,
    reaching_time TIME,
    star_rating FLOAT,
    price DECIMAL(10, 2),
    seats_available INT
);
'''
cursor.execute(create_table_query)
conn.commit()
```

```python
# insert data into database
insert_query = '''
INSERT INTO redbus_data (route_name, route_link, busname, bustype, departing_time, 
                             duration, reaching_time, star_rating, price, seats_available) 
VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
'''
data = df.values.tolist()
cursor.executemany(insert_query, data)
conn.commit()

print('Successfully inserted data into database')
```
